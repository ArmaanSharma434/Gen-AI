```python
def pass_students(**kwargs):
  pass_student_list= []
  for name, marks in kwargs.items():
    if marks>35:
      pass_student_list.append(name)
  return pass_student_list
```

```python
pass_students(A = 30, B = 40, C = 50, D = 80, E = 20)
```

```python
import time

lyrics = [
    "Sukoon kyu ye dil kyu pata nhi",
    "Kyu todu khud se jo the wade",
    "Ke ab ye ishq nibhana nhi",
    "Main modun tum se jo ye chehra",
    "Dubara nazar milana nhi!"
]

delays = [2.0, 1.8, 2.1, 2.4, 1.7]

print("The song tonight:")
time.sleep(1.5)

for i, line in enumerate(lyrics):
  print(line)
  time.sleep(delays[i])
```

```python
import time

lyrics = [
     "Sukoon kyu ye dil kyu pata nhi",
    "Kyu todu khud se jo the wade",
    "Ke ab ye ishq nibhana nhi",
    "Main modun tum se jo ye chehra",
    "Dubara nazar milana nhi!"
]

word_delay = 0.3 #delay btw words
line_delay = 1.5 #delay btw lines

print("The song tonight:")
time.sleep(1.5)
#loop through each line
for line in lyrics:
  words = line.split() #splits line into words

  for word in words:
    print(word, end="", flush=True)
    time.sleep(word_delay)
    print()

  time.sleep(line_delay)
```

```python
from transformers import pipeline

summarizer  = pipeline("summarization", model="facebook/bart-large-cnn")

text = """The global shift toward renewable energy is accelerating, driven by concerns over climate change and energy security. Solar
 and wind power installations have reached record levels worldwide."""

summary = summarizer(text, max_length=30, min_length=10, do_sample=False)
print(summary[0])
```
